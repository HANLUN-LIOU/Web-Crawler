```python
import requests
import json
from bs4 import BeautifulSoup

# Step 1: Get entire page from cwa and save as sample.html
url = "https://opendata.cwa.gov.tw/fileapi/v1/opendataapi/F-A0010-001?Authorization=CWA-2DAD4A1F-88FA-4364-AD01-308D10FBAE53&downloadType=WEB&format=JSON"
response = requests.get(url)
if response.status_code == 200:
    with open("sample.html", "wb") as file:
        file.write(response.content)
    print("Sample HTML saved as sample.html")
else:
    print("Failed to fetch data from the API")

# # Step 2: Read the template HTML and extract information from sample.html
# with open("template.html", "r") as file:
#     template_html = file.read()

# # Parse the JSON string into a Python dictionary
# template_data = json.loads(template_html)

# Read the sample HTML
with open("sample.html", "r") as file:
    sample_html = file.read()

# Parse the sample HTML and extract necessary information
soup = BeautifulSoup(sample_html, 'html.parser')
regions = ["北部地區", "中部地區", "南部地區", "東北地區", "東部地區", "東南地區"]
weather_data = {}

for region in regions:
    weather_data[region] = {}
    wx_data = []
    maxt_data = []
    mint_data = []

    wx_tags = soup.select(f'locationName:contains("{region}") + .weatherElements > .Wx > .daily')
    maxt_tags = soup.select(f'locationName:contains("{region}") + .weatherElements > .MaxT > .daily')
    mint_tags = soup.select(f'locationName:contains("{region}") + .weatherElements > .MinT > .daily')

    for wx_tag, maxt_tag, mint_tag in zip(wx_tags, maxt_tags, mint_tags):
        wx_data.append({
            "dataDate": wx_tag.select_one(".dataDate").text.strip(),
            "weather": wx_tag.select_one(".weather").text.strip(),
            "weatherid": wx_tag.select_one(".weatherid").text.strip()
        })
        maxt_data.append({
            "dataDate": maxt_tag.select_one(".dataDate").text.strip(),
            "temperature": maxt_tag.select_one(".temperature").text.strip()
        })
        mint_data.append({
            "dataDate": mint_tag.select_one(".dataDate").text.strip(),
            "temperature": mint_tag.select_one(".temperature").text.strip()
        })

    weather_data[region]["Wx"] = wx_data
    weather_data[region]["MaxT"] = maxt_data
    weather_data[region]["MinT"] = mint_data

# Save the extracted weather data as JSON
with open("weather_data.json", "w") as file:
    json.dump(weather_data, file)

print("Weather data extracted and saved as weather_data.json")
```

```python
import json

# Read the JSON data
with open("sample.html", "r") as file:
    json_data = file.read()

# Parse the JSON string into a Python dictionary
data = json.loads(json_data)

# Extract temperature information for each location
for location in data['agrWeatherForecasts']['weatherForecasts']['location']:
    location_name = location['locationName']
    max_temps = [int(day['temperature']) for day in location['weatherElements']['MaxT']['daily']]
    min_temps = [int(day['temperature']) for day in location['weatherElements']['MinT']['daily']]
    # Assuming the data is for a week
    assert len(max_temps) == len(min_temps) == 7, "Data doesn't cover a week"
    # Print everyday max and min temperature
    for i, (max_temp, min_temp) in enumerate(zip(max_temps, min_temps), 1):
        print(f"{location_name}: Day {i}, 最高溫度 {max_temp}°C, 最低溫度 {min_temp}°C")
```
